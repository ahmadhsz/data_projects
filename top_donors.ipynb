## Top Most Donors To Afghanistan
### Source: https://stats.oecd.org

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Technical Cooperation Grants

```python
tech_df = pd.read_csv('technical_cooperation.csv')
tech_df = tech_df.drop(tech_df.loc[tech_df['Donor'] == 'EU Institutions'].index)
tech_df = tech_df[['Donor', 'Year', 'Value']]
tech_df = tech_df.set_index('Donor')
tech_df.head()
```

```python
tech_donors = tech_df.groupby('Donor').sum('Value').sort_values(by='Value', ascending=False).head().drop('Year', axis=1)
tech_donors
```

```python
fig = plt.figure(figsize=(20, 10), dpi=150)
ax = sns.barplot(x=tech_donors.index, y='Value',
            data=tech_donors,
            palette='magma',
            order=tech_donors.index)
sns.set(font_scale=1.3)
ax.bar_label(ax.containers[0], fmt='%.2f' )
plt.ylabel('Value In Million ($)')
plt.xlabel('Donors')
plt.title('Total Value From 2001-2020 (Technical Cooperation Grants)')
#plt.savefig('1.1_total_technical.jpeg')
```

```python
tech_top_5 = tech_df[tech_df.index.isin(tech_donors.index)]
tech_top_5
```

```python
fig = plt.figure(figsize=(20, 10), dpi=150)
sns.barplot(x=tech_top_5.index, y='Value',
            data=tech_top_5,
            hue='Year',
            palette='Paired',
            order=tech_donors.index)
sns.set(font_scale=1.3)
plt.ylabel('Value In Million ($)')
plt.xlabel('Donors')
plt.title('Technical Cooperation Top 5 Donors To Afghanistan')
#plt.savefig('1.2_tech_top_5.jpeg')
```

```python
tech_top_5.to_csv('technical_cooperation_top_5.csv')
```

# Germany

### let's focus on Germany's stats

```python
ger_df = pd.read_csv('germany.csv')
ger_df = ger_df[ger_df['Donor'] == 'Germany']
ger_df = ger_df[['Donor', 'Series', 'Year', 'Value']]
ger_df
```

```python
series = ger_df['Series'].unique().tolist()
averages = {}
for item in series:
    mean = ger_df[ger_df['Series'] == item]['Value'].mean()
    averages[item] = mean
x = pd.DataFrame.from_dict(averages, orient='index', columns=['average'])
x.average = x.average.round(2)
x.to_csv('series_avg.csv')
```

## ODA By Sector

```python
by_sector = ger_df.groupby('Series').sum('Value').sort_values(by='Value', ascending=False).drop('Year', axis=1)
by_sector
```

```python
fig = plt.figure(figsize=(22, 12), dpi=150)
ax = sns.barplot(y=by_sector.index, x='Value',
                data=by_sector,
                palette='magma')
sns.set(font_scale=1.3)
ax.bar_label(ax.containers[0], fmt='%.2f' )
plt.xlabel('Value In Million ($)')
plt.ylabel('Sector')
plt.title('Total Value From 2001-2020 By Sector (Germany To Afghanistan)')
plt.savefig('5.1_by_sector.jpeg')
```

```python
fig = plt.figure(figsize=(25, 10), dpi=250)
ax = sns.barplot(x='Year', y='Value',
                data=ger_df,
                hue='Series',
                palette='Paired')
sns.set(font_scale=1.3)
#ax.bar_label(ax.containers[0], fmt='%.2f' )
plt.xlabel('Value In Million ($)')
plt.ylabel('Sector')
plt.title('Total Value From 2001-2020 By Sector And Year (Germany To Afghanistan)')
#plt.savefig('5.2_by_sector_year.jpeg')
```
